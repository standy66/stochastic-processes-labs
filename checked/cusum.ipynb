```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
def cusum(sample, pdf_inf, pdf_0):
    s = [0]
    for x in sample:
        s.append(max(s[-1] + np.log(pdf_0(x) / pdf_inf(x)), 0))
    return np.array(s)

def exponential_smoothing(sample, alpha=0.5):
    smoothing = [np.median(sample[:1000])]
    for i in range(1, len(sample)):
        smoothing.append((1 - alpha) * smoothing[-1] + alpha * sample[i])
    return np.array(smoothing)
```

# 1

```python
N = 10 ** 4
s_loc = 0.5
s_scale = 0.02
new_loc = 0.57
new_scale = 0.03
sample = np.zeros(N)
theta = sps.randint.rvs(low=1500, high=N - 1500)
print 'theta =', theta

sample[:theta] = sps.norm.rvs(size=theta, loc=s_loc, scale=s_scale)
ind = sps.bernoulli.rvs(size=N - theta, p=0.9)
sample[theta:] = ind * sps.norm.rvs(size=N - theta, loc=s_loc, scale=s_scale) \
               + (1 - ind) * sps.norm.rvs(size=N - theta, loc=new_loc, scale=new_scale)

plt.figure(figsize=(20, 2))
plt.plot(sample)
plt.show()

plt.figure(figsize=(20, 2))
plt.plot(exponential_smoothing(sample, alpha=0.001))
plt.show()

indexs_hard = sps.bernoulli.rvs(size=N, p=0.01)
hard_sample = indexs_hard * sps.expon.rvs(size=N, loc=s_loc, scale=0.05) \
               + (1 - indexs_hard) * sample
    
indexs_hard = sps.bernoulli.rvs(size=N, p=0.01)
hard_sample = indexs_hard * sps.expon.rvs(size=N, loc=s_loc, scale=0.02) \
               + (1 - indexs_hard) * hard_sample
    
plt.figure(figsize=(20, 2))
plt.plot(hard_sample)
plt.show()

plt.figure(figsize=(20, 2))
plt.plot(exponential_smoothing(hard_sample, alpha=0.001))
plt.show()

plt.figure()
plt.plot(cusum(hard_sample, sps.norm(loc=0.5, scale=0.01).pdf, sps.norm(loc=0.54, scale=0.04).pdf))
plt.show()

plt.figure()
plt.plot(cusum(exponential_smoothing(hard_sample, alpha=0.001),
               sps.norm(loc=0.50, scale=0.001).pdf, sps.norm(loc=0.506, scale=0.002).pdf))
plt.show()

np.savetxt('./task 3/1. times 1.txt', hard_sample, fmt='%.7f')
np.savetxt('./task 3/1. theta 1.txt', [theta], fmt='%d')
```
